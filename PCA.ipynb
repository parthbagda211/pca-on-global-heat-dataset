# Cleaning Data

```python
# Cleaning up the data
clean1 = data[data['Hour'] < 6]
clean2 = data[data['Hour'] > 18]

print("Cleaning Data from 12:00 am to 5:59 am")
clean1.head(10)
```

```python
print("Cleaning Data from 18:00 pm to  12.00 am")
clean2.head(10)
```

```python
data =  data.drop(clean1.index, axis=0)
data =  data.drop(clean2.index, axis=0)
```

```python
print("Cleaned Data Description : ")
data.describe()
```

```python
print("Cleaned Data : ")
data.head(5)
```

```python
data.shape
```

```python
data = data[0:1000]
```

# Preparing Training and Testing Data

```python
x = data.iloc[:, 4:10].values
y = data.iloc[:, 10].values 

y = y.reshape(-1, 1)

from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(x, y, test_size = 0.2, random_state = 0)
```

```python
print (x_train)
```

```python
print (y_train)
```

# Standardizing DATA

```python
from sklearn.preprocessing import StandardScaler
sc_X = StandardScaler()
x_train = sc_X.fit_transform(x_train)
x_test = sc_X.transform(x_test)
sc_y = StandardScaler()
y_train = sc_y.fit_transform(y_train)
```

```python
print (x_train)
```

```python
print (y_train)
```

# <font color = "green">K Means with 6 clusters</font>

```python
from sklearn.cluster import KMeans

kmeans  = KMeans(n_clusters = 6, random_state = 0, n_jobs=-1)
kmeans = kmeans.fit(x_train)
```

```python
# View predict class
labels = kmeans.labels_
print("Labels : ", labels)

# View cluster centers
cluster_centers = kmeans.cluster_centers_
print("\n\n Cluster Centroids : \n", cluster_centers)
```

# <font color = "green">Applying PCA - Dimensional Reduction </font>

```python
from sklearn.decomposition import PCA

pca = PCA(n_components=2).fit(x_train)
pca = pca.transform(x_train)
```

```python
pca
```

```python
x_train
```

## Visualisation

```python
for i in range(0, pca.shape[0]):
    if kmeans.labels_[i] == 0:
        c1_6 = plt.scatter(pca[i,0],pca[i,1],c='orange', marker='^')
    
    elif kmeans.labels_[i] == 1:
        c2_6 = plt.scatter(pca[i,0],pca[i,1],c='g', marker='o')
    
    elif kmeans.labels_[i] == 2:
        c3_6 = plt.scatter(pca[i,0],pca[i,1],c='b', marker='*')
    
    elif kmeans.labels_[i] == 3:
        c4_6 = plt.scatter(pca[i,0],pca[i,1],c='yellow', marker='*')
  
    elif kmeans.labels_[i] == 4:
        c5_6 = plt.scatter(pca[i,0],pca[i,1],c='cyan', marker='>')
    
    elif kmeans.labels_[i] == 5:
        c6_6 = plt.scatter(pca[i,0],pca[i,1],c='red', marker='+')
    
plt.rcParams["figure.figsize"] = (12, 10)
plt.legend([c1_6, c2_6, c3_6, c4_6, c5_6, c6_6],['Cluster 0', 'Cluster 1', 
                                                 'Cluster 2', 'Cluster 3', 
                                                 'Cluster 4', 'Cluster 5'])
plt.title('K-means clustering with 6 clusters')
plt.show()
```

```python
from sklearn.cluster import AgglomerativeClustering

aglo_clt_4 = AgglomerativeClustering(linkage='complete', affinity='euclidean', n_clusters=4)

# Train model
model_aglo_4 = aglo_clt_4.fit(x_train)

# View predict class
model_4_aglo_labels = model_aglo_4.labels_

print("Labels : ", model_aglo_4.labels_)
```

```python
# Applying Dimensional Reduction
pca_4_aglo = PCA(n_components=2).fit(x_train)
pca_4_aglod_4_aglo = pca_4_aglo.transform(x_train)

# Plotting
for i in range(0, pca_4_aglod_4_aglo.shape[0]):
    
    if model_aglo_4.labels_[i] == 0:
        c1_4_aglo = plt.scatter(pca_4_aglod_4_aglo[i,0],pca_4_aglod_4_aglo[i,1],c='orange', marker='^')
    
    elif model_aglo_4.labels_[i] == 1:
        c2_4_aglo = plt.scatter(pca_4_aglod_4_aglo[i,0],pca_4_aglod_4_aglo[i,1],c='g', marker='o')
    
    elif model_aglo_4.labels_[i] == 2:
        c3_4_aglo = plt.scatter(pca_4_aglod_4_aglo[i,0],pca_4_aglod_4_aglo[i,1],c='r', marker='*')
    
    elif model_aglo_4.labels_[i] == 3:
        c4_4_aglo = plt.scatter(pca_4_aglod_4_aglo[i,0],pca_4_aglod_4_aglo[i,1],c='blue', marker='+')   
    

plt.rcParams["figure.figsize"] = (12, 10)
plt.legend([c1_4_aglo, c2_4_aglo, c3_4_aglo, c4_4_aglo],['Cluster 0', 'Cluster 1', 'Cluster 2', 'Cluster 3'])
plt.title('Agglomerative clustering with 4 clusters')
plt.show()
```
